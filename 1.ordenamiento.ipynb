# 1. ordenamiento de una cadena de n√∫meros

### Este ejercicio conciste en reacomodar los datos de un vector, los primeros los ultimo.

```python
def reordenamiento(vector):
  vector_ro = list()
  for i in range(len(vector)):
    vector_ro.append(vector[(len(vector)-1)-i])
  return vector_ro
```

```python
reordenamiento(['A',8,9,7,85,2])
```
